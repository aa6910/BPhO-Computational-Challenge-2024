## Tensor broadcasting

Sometimes you want to e.g. add a matrix to a vector (adding the vector to each of the matrix, perhaps). When we try to add / multiply / whatever two tensors with different shapes, Pytorch will attempt to automatically "broadcast" one tensor across the other. Broadcasting can get tricky, but it is very useful, as code written using broadcasting will run much faster than doing the same operation using python for-loops.

```python
#You can do this, which will add the vector y to each row of X
X = t.ones(3,4)
y = t.randn(4)
print(X)
print(y)
print(X+y)
```

```python
#But you can't do this:
X = t.ones(3,4)
y = t.rand(3)
print(X+y)
```

To understand what works / doesn't, we need to understand the **broadcasting rules**!

From the [Pytorch documentation](https://pytorch.org/docs/stable/notes/broadcasting.html):

*Two tensors are “broadcastable” if the following rules hold:*
- *Each tensor has at least one dimension.*
- *When iterating over the dimension sizes, starting at the trailing dimension, the dimension sizes must either be equal, one of them is 1, or one of them does not exist.*

Let's try and understand these in code.

```python
#Key rule: dimensions must match, or be 1:
X = t.ones(2,3,4)
Y = t.rand(1,3,1)
#Here, the middle dimensions match.
#The first dimensions are 2 (for X) and 1 (for Y), so the Y-dimension broadcasts
#The third dimensions are 4 (for X) and 1 (for Y), so again, the Y-dimension broadcasts.
Z = X+Y
print(Z.shape)
```

This translates into maths as:
\begin{align}
  Z_{ijk} = X_{ijk} + Y_j
\end{align}

Note that we only have dimension size $1$ for $Y$ in the $i$ and $k$ dimensions, so it's as if those indices don't exist on the Y tensor.

In other words, since the $j$ dimensions of $X$ and $Y$ match, and $Y$ only has size 1 in dimensions $i$ and $k$, then we can add the $j$ dimensions together for all $i$ and for all $k$. If $X$, which is of size $(2,3,4)$, were viewed as a list of 2 matrices of size $3 \times 4$, then we're adding the vector $Y$ of length 3 to every column of every matrix in $X$ (since each column of each matrix is a vector of length 3, this makes sense).

Without broadcasting, we'd need to write a python for-loop to achieve something like this, which, for very big tensors, will be very slow. Broadcasting uses fast C code, and is therefore much quicker.

What happens if the tensors have different numbers of dimensions?

```python
#Here, X is a 3-dimensional tensor, while Y is a 1-dimensional tensor,
X = t.ones(2,3,4)
Y = t.ones(4)
Z = X+Y
print(Z.shape)
```

This expression translates to math as,
\begin{align}
  Z_{ijk} = X_{ijk} + Y_k.
\end{align}
There are two ways to think about this:
* PyTorch matches dimensions from right to left,
* PyTorch fills the "missing" dimensions on the left with ones.  In which case, the following code is equivalent,

```python
#The above code works is equivalent to this
X = t.ones(2,3,4)
Y = t.ones(1,1,4)
Z = X+Y
print(Z.shape)
```

However, the following code gives an error, because `X.shape = torch.Size([2])` behaves like `Y.shape = torch.Size([1,1,2])`, and this doesn't broadcast with `X.shape = torch.Size(2,3,4)`, because in the third dimesion, $2 \neq 4$.

```python
# This code gives an error
X = t.ones(2,3,4)
Y = t.rand(2)
X+Y
```

This expression translates to math in the same way, as
\begin{align}
  Z_{ijk} = X_{ijk} + Y_k.
\end{align}
But now the $k$ dimension on X (size 4) and the $k$ dimension on Y (size 2) don't match, which gives an error.

```python
#The above code behaves just like, and therefore gives us an error
X = t.ones(2,3,4)
Y = t.rand(1,1,2)
X+Y
```

But what if we do want to do broadcast along the first dimension?  We need to construct $Y$ so that the length 2 dimension appears in the right place!

```python
# Manually inserting ones so we can sum the first dimensions of X and y together
X = t.ones(2,3,4)
Y = t.randn(2,1,1)
Z = X+Y
print(X)
print(Y)
print(Z)
print(Z.shape)
```

This expression translates to math as
\begin{align}
  Z_{ijk} = X_{ijk} + Y_i.
\end{align}
So now the "i" dimension on X and the "i" dimension on Y do match! From the output of the above code, we can see this is like $X$ being 2 matrices of size $3 \times 4$, and we add the first element of $Y$ to the first matrix, and the second element of $Y$ to the second matrix.

But what if we already have a tensor Y, but it's the wrong shape (doesn't have the extra 1s needed)?

```python
X = t.ones(2,3,4)
Y = t.randn(2)
#Y is the wrong shape to broadcast with X!
print(Y.shape) #torch.Size([2])
# and so X+Y gives an error:
X+Y
```

We can add extra dimensions to Y using `Y[:, None, None]` (this is special Pytorch notation, don't attach any significance to the word "None" here)

```python
Y = Y[:, None, None]
#The None's add extra 1 dimensions, we have a tensor of the right shape!
print(Y.shape) #torch.Size([2,1,1])
#And the broadcasting works,
print(X+Y)
```

# Additional useful operations

```python
#Matrix operations
A = t.randn((2,3))
print("Matrix")
print(A)
print("\nMatrix Transpose")
print(A.mT)
print("\nElementwise product")
print(A * A)
print("\nMatrix Product")
print(A @ A.mT)
```

```python
#Matrix-vector operations
N = 2
M = 3
A = t.randn((M, N))
a_vec = t.randn(M)
b_vec = t.randn(N)

#Interprets a_vec as a row-vector (otherwise matrix-vector product wouldn't make sense)
#Input is a vector (with one dimension); output is a vector (with one dimension)
print((a_vec @ A).shape)
#Interprets b_vec as a column-vector (otherwise matrix-vector product wouldn't make sense)
#Input is a vector (with one dimension); output is a vector (with one dimension)
print((A @ b_vec).shape)
#We can also convert a_vec to an explicit row-vector, by converting to a matrix
#of shape 1xM
a_rowvec = a_vec[None, :]
print(a_rowvec)
#Now, input is a row-vector in matrix form (1xM), so output is a row-vector in matrix form (1xN)
print((a_rowvec @ A).shape)
#Likewise, we can convert b_vec to an explicit column-vector, by converting to a matrix
#of shape Nx1
b_colvec = b_vec[:, None]
print(b_colvec)
#Now, input is a row-vector in matrix form (Nx1), so output is a row-vector in matrix form (Mx1)
print((A @ b_colvec).shape)

```

##Reduction operations

There are a bunch of operations (mean, sum, prod, std, var) etc.
They all do the obvious things.
But the key trick is understanding how these operations interact with Tensor shape.

```python
A = t.randn(2,3,4)
```

```python
#All these ops reduce over a dimension, and we need to provide an explicit index for that dimension.
print(A.mean(0).shape)
print(A.mean(1).shape)
print(A.mean(2).shape)
#We can index from the back, as in list/tuple indexing
print(A.mean(-1).shape) # In this case, this is equivalent to A.mean(2) because 2 is the last dimension (dimensions are zero indexed, 0,1,2)
```

`A.mean(0)` translates to math as:
\begin{align}
M_{jk} = \tfrac{1}{2} \sum_{i=1}^2 A_{ijk}
\end{align}
`A.mean(1)` translates to math as:
\begin{align}
M_{ik} = \tfrac{1}{3} \sum_{j=1}^3 A_{ijk}
\end{align}
`A.mean(2)` translates to math as:
\begin{align}
M_{ij} = \tfrac{1}{4} \sum_{k=1}^4 A_{ijk}
\end{align}

What if we want to mean-subtract $A$?  That's easy if we want to take the mean over the first dimension, as the shapes broadcast:

```python
# A and A_mean follow the broadcasting rules in this case
A_mean = A.mean(0)
print(f"A.shape =   {A.shape}")
print(f"A_mean.shape = {A_mean.shape}")
A-A_mean
```

But mean-subtracting $A$ doesn't automatically work if we want to take the mean over the second or third dimension, as the resulting mean Tensor doesn't broadcast with $A$.

```python
# A and A_mean don't follow the broadcasting rules in this case, so will throw an error
A_mean = A.mean(2)
print(f"A.shape =   {A.shape}")
print(f"A_mean.shape = {A_mean.shape}")
A-A_mean
```

We could fix this using `None` to add the extra 1 back in:

```python
# Using None to add 1s back in will work
A_mean = A.mean(2)[:, :, None]
print(f"A.shape =   {A.shape}")
print(f"A_mean.shape = {A_mean.shape}")
A-A_mean
```

But this is tedious, requiring us to make sure we put `None` in the correct place that we took the mean over. Looking at [the docs](https://pytorch.org/docs/stable/generated/torch.mean.html), there's an easy way to fix this bug: use the `keepdim=True` kwarg!
This keeps a single element in the final dimension, so broadcasting still works!

```python
# Using the keepdim=True kwarg makes thing simpler.
A_mean = A.mean(2, keepdim=True)
print(f"A.shape =      {A.shape}")
print(f"A_mean.shape = {A_mean.shape}")
A-A_mean
```
